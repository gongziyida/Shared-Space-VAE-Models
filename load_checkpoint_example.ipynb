```python
import os
import torch
import matplotlib.pyplot as plt
from mnistsvhn.experiment import MNISTSVHN
```

```python
EXP_DIR = 'runs/MNIST_SVHN/joint_elbo/non_factorized/laplace_laplace/SVHN_MNIST_2022_01_11_22_05_07_526073/'
```

```python
flags = torch.load(os.path.join(EXP_DIR, 'flags.rar'))
```

```python
exp = MNISTSVHN(flags, [])
```

```python
vae = exp.mm_vae
```

```python
vae.load_state_dict(torch.load(os.path.join(EXP_DIR, 'checkpoints/0009/mm_vae')))
```

```python
dataset = exp.dataset_test
test_loader = torch.utils.data.DataLoader(dataset, batch_size=20, shuffle=True)
```

```python
for (data, label) in test_loader:
    for key in data.keys():
        data[key] = data[key].to(flags.device)
    results = vae(data)
    break
```

```python
rec_mnist = results['rec']['mnist'].mean
rec_svhn = results['rec']['svhn'].mean
```

```python
ncols = min(rec_mnist.shape[0], 20)
fig, ax = plt.subplots(nrows=2, ncols=ncols, figsize=(15, 1.5))
for i, aux in enumerate(zip(data['mnist'], rec_mnist)):
    if i >= ncols:
        break
    for j, im in enumerate(aux):
        ax[j, i].imshow(im.detach().cpu().numpy().reshape(28, 28), cmap='gray')
        ax[j, i].set_xticks([])
        ax[j, i].set_yticks([])
ax[1, 0].set_ylabel('Rec', fontsize=24)
fig.tight_layout(pad=0)
```

```python
ncols = min(rec_svhn.shape[0], 20)
fig, ax = plt.subplots(nrows=2, ncols=ncols, figsize=(15, 1.5))
for i, aux in enumerate(zip(data['svhn'], rec_svhn)):
    if i >= ncols:
        break
    for j, im in enumerate(aux):
        ax[j, i].imshow(im.detach().cpu().numpy().transpose(1, 2, 0))
        ax[j, i].set_xticks([])
        ax[j, i].set_yticks([])
ax[1, 0].set_ylabel('Rec', fontsize=24)
fig.tight_layout(pad=0)
```
